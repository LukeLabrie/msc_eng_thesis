```python
import matplotlib.pyplot as plt
from parameters import *
import numpy as np
```

```python
# unpack data 
filename1 = 'data/sim_out_999.99_0.2.txt' # 1.2 MW
filename2 = 'data/sim_out_999.99_0.3.txt' # 1.3 MW
filename3 = 'data/sim_out_999.99_1.0.txt' # 1.1 MW
filename4 = 'data/sim_out_999.99_2.14.txt' # 2.14 MW
filename5 = 'data/sim_out_999.99_5.0.txt' # 5.0 MW


sol_jit1 = []
sol_jit2 = []
sol_jit3 = []
sol_jit4 = []
sol_jit5 = []

# Open the file for reading
with open(filename1, 'r') as file:
    # Iterate over each line in the file
    for line in file:
        # Split the line into components, convert each to float, and append to sol_jit
        sol_jit1.append([float(value) for value in line.split()])
# Open the file for reading
with open(filename2, 'r') as file:
    # Iterate over each line in the file
    for line in file:
        # Split the line into components, convert each to float, and append to sol_jit
        sol_jit2.append([float(value) for value in line.split()])
# Open the file for reading
with open(filename3, 'r') as file:
    # Iterate over each line in the file
    for line in file:
        # Split the line into components, convert each to float, and append to sol_jit
        sol_jit3.append([float(value) for value in line.split()])
# Open the file for reading
with open(filename4, 'r') as file:
    # Iterate over each line in the file
    for line in file:
        # Split the line into components, convert each to float, and append to sol_jit
        sol_jit4.append([float(value) for value in line.split()])
# Open the file for reading
with open(filename5, 'r') as file:
    # Iterate over each line in the file
    for line in file:
        # Split the line into components, convert each to float, and append to sol_jit
        sol_jit5.append([float(value) for value in line.split()])

sol_jit1 = [sol_jit1[s] for s in range(len(sol_jit1)-1)]
sol_jit2 = [sol_jit2[s] for s in range(len(sol_jit2)-1)]
sol_jit3 = [sol_jit3[s] for s in range(len(sol_jit3)-1)]
sol_jit4 = [sol_jit4[s] for s in range(len(sol_jit4)-1)]
sol_jit5 = [sol_jit5[s] for s in range(len(sol_jit5)-1)]
```

```python
# timesteps
T1 = [s[0] for s in sol_jit1]
T2 = [s[0] for s in sol_jit2]
T3 = [s[0] for s in sol_jit3]
T4 = [s[0] for s in sol_jit4]
T5 = [s[0] for s in sol_jit5]

# total power
tot_pow1 = [2*k_f1*0.2*s[27] for s in sol_jit1]
tot_pow2 = [2*k_f1*0.3*s[27] for s in sol_jit2]
tot_pow3 = [2*k_f1*1.0*s[27] for s in sol_jit3]
tot_pow4 = [2*k_f1*2.14*s[27] for s in sol_jit4]
tot_pow5 = [2*k_f1*5.0*s[27] for s in sol_jit5]

# heat extracted 1
fh_ex1 = [(s[2]-s[1])*W_f*scp_f for s in sol_jit1]
ch_ex1 = [(s[5]-s[4])*W_c*scp_c for s in sol_jit1]
hwf_ex1 = [(s[11]-s[10])*W_h_fh*scp_h for s in sol_jit1]
hwc_ex1 = [(s[16]-s[15])*W_h_ch*scp_h for s in sol_jit1]
tot_ex1 = [fh_ex1[i]+ch_ex1[i]+hwf_ex1[i]+hwc_ex1[i] for i in range(len(sol_jit1))]

# heat extracted 2
fh_ex2 = [(s[2]-s[1])*W_f*scp_f for s in sol_jit2]
ch_ex2 = [(s[5]-s[4])*W_c*scp_c for s in sol_jit2]
hwf_ex2 = [(s[11]-s[10])*W_h_fh*scp_h for s in sol_jit2]
hwc_ex2 = [(s[16]-s[15])*W_h_ch*scp_h for s in sol_jit2]
tot_ex2 = [fh_ex2[i]+ch_ex2[i]+hwf_ex2[i]+hwc_ex2[i] for i in range(len(sol_jit2))]

# heat extracted 3
fh_ex3 = [(s[2]-s[1])*W_f*scp_f for s in sol_jit3]
ch_ex3 = [(s[5]-s[4])*W_c*scp_c for s in sol_jit3]
hwf_ex3 = [(s[11]-s[10])*W_h_fh*scp_h for s in sol_jit3]
hwc_ex3 = [(s[16]-s[15])*W_h_ch*scp_h for s in sol_jit3]
tot_ex3 = [fh_ex3[i]+ch_ex3[i]+hwf_ex3[i]+hwc_ex3[i] for i in range(len(sol_jit3))]

# heat extracted 4
fh_ex4 = [(s[2]-s[1])*W_f*scp_f for s in sol_jit4]
ch_ex4 = [(s[5]-s[4])*W_c*scp_c for s in sol_jit4]
hwf_ex4 = [(s[11]-s[10])*W_h_fh*scp_h for s in sol_jit4]
hwc_ex4 = [(s[16]-s[15])*W_h_ch*scp_h for s in sol_jit4]
tot_ex4 = [fh_ex4[i]+ch_ex4[i]+hwf_ex4[i]+hwc_ex4[i] for i in range(len(sol_jit4))]

# heat extracted 5
fh_ex5 = [(s[2]-s[1])*W_f*scp_f for s in sol_jit5]
ch_ex5 = [(s[5]-s[4])*W_c*scp_c for s in sol_jit5]
hwf_ex5 = [(s[11]-s[10])*W_h_fh*scp_h for s in sol_jit5]
hwc_ex5 = [(s[16]-s[15])*W_h_ch*scp_h for s in sol_jit5]
tot_ex5 = [fh_ex5[i]+ch_ex5[i]+hwf_ex5[i]+hwc_ex5[i] for i in range(len(sol_jit5))]
```
